```python
!pip install transformers

from transformers import pipeline

# Load text understanding model
text_classifier = pipeline("sentiment-analysis")

# Load text generation model
text_generator = pipeline("text-generation")

# Define input text
input_text = "I loved the movie. It was fantastic!"

# Perform sentiment analysis on input text
sentiment = text_classifier(input_text)[0]

# Generate response based on sentiment
if sentiment['label'] == 'POSITIVE':
    response = text_generator("Tell me more about why you loved the movie.", max_length=50)[0]['generated_text']
else:
    response = text_generator("I'm sorry to hear that. What didn't you like about the movie?", max_length=50)[0]['generated_text']

# Print response
print(response)
```
